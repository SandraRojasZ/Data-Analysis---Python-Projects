## Renda per capita
A renda per capita mede a renda de cada indivíduo dentro de uma determinada população, calculando uma média geral desse valor.<p>
Renda Per capita = PNB / numero total de habitantes

```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('Dados_Pib.xlsx')
```

```python
df = df[['Territorialidades','Ano','PIB per capita']]
```

```python
df.head(10)
```

```python
df.groupby(by=['Territorialidades','Ano']).mean()
```

```python
# Sistema de Grids

# Cor de Fundo
cor_fundo = '#f5f5f5'

#criar o sistema de grids
grif_graficos = sns.FacetGrid(df, col = 'Territorialidades', hue = 'Territorialidades', col_wrap = 4);

# adicionar grafico linhas em cada gráfico plotado
grif_graficos = grif_graficos.map(plt.plot, 'Ano', 'PIB per capita');

# adicionar sombra + ajuste do titulo
grif_graficos = grif_graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha = 0.2). set_titles('{col_name} Territorialidades');

#filtrar o titulo
grif_graficos = grif_graficos.set_titles('{col_name}')

#adicionar um subtitulo
grif_graficos = grif_graficos.fig.suptitle(
    'Evolução da Renda per capita por Estado \n Esse relatório foi elaborado no treinamento "Python para análise de Dados"',
    fontsize=18
)

#Ajustando o Titulo
plt.subplots_adjust( top = 0.92)
```
