## Empresas de Energia na B3

O objetivo dessa análise é comparar as ações das empresas de energia X PETROBRAS.

```python
# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para análises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#lendo as base de dados
df = pd.read_excel('Dados_Empresas_Energia.xlsx')
df.head()
```

```python
# Serie temporais - Setar o Index
df.set_index('Data', inplace=True)
```

```python
df.head()
```

```python
#Grafico

# Estilo do gráfico
plt.style.use('seaborn-darkgrid')

#Tamanho
plt.figure(figsize = (16,7))

#Titulo
plt.title('Analises ações de empresas de Energia', fontsize = 18, fontweight = 0)

#Plot da Petrobras
plt.plot(df.index, df['Petrobras'], color = '#008c4a', linewidth=4, alpha = 0.7)
#Texto Petrobras
plt.text(df.index[-1], df['Petrobras'].tail(1), 'Petrobras', color = '#008c4a', size = 'large', horizontalalignment = 'left')

#Plot de todas as colunas
for coluna in df.columns[1:]:
    #Plot das outras ações
    plt.plot(df.index, df[coluna], color = 'gray', linewidth=3, alpha = 0.7)
    #Texto das outras colunas
    plt.text(df.index[-1], df[coluna].tail(1), coluna, color = 'gray', size = 'large', horizontalalignment = 'left')


#labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
